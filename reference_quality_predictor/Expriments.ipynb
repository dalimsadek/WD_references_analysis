```python

import pandas as pd
import re

# Simulated raw text content
with open("all_model_results.txt", "r") as f:
    content = f.read()

# Split content into blocks
blocks = content.split("Model: ")[1:]  # First split will be empty

rows = []
for block in blocks:
    lines = block.strip().splitlines()
    model_line = lines[0]
    task_line = lines[1]
    result_line = lines[2] if len(lines) > 2 else ""

    model = model_line.strip()
    task = task_line.replace("Task: ", "").strip()

    if "Could not parse output" in result_line or result_line.strip() == "":
        rows.append([model, task, "", "", "", "", "", ""])
    else:
        result_metrics = re.findall(r"([\w_]+):([\d\.eE\+\-]+)", result_line)
        metrics_dict = dict(result_metrics)
        rows.append([
            model,
            task,
            metrics_dict.get("precision", ""),
            metrics_dict.get("recall", ""),
            metrics_dict.get("f1", ""),
            metrics_dict.get("auc_pr", ""),
            metrics_dict.get("mcc", ""),
            metrics_dict.get("f1_new", "")
        ])

df = pd.DataFrame(rows, columns=["Model", "Task", "Precision", "Recall", "F1", "AUC_PR", "MCC", "F1_NEW"])
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('results/prediction_data.csv')  # Or your full feature file

# Map label values
df['is_relevant'] = df['support_object'].apply(lambda x: x == 0)
df['is_authoritative'] = df['authoritative'].apply(lambda x: x == 0)

# Define user groups
user_groups = {
    'Humans': df[df['user_type'] == 1],
    'Bots': df[df['user_type'] == 2],
    'All Users': df
}

# Prepare results
result = {}

for group_name, group_df in user_groups.items():
    total = len(group_df)
    result[group_name] = {
        'Relevant & authoritative': len(group_df[(group_df['is_relevant']) & (group_df['is_authoritative'])]) / total * 100,
        'Relevant & not authoritative': len(group_df[(group_df['is_relevant']) & (~group_df['is_authoritative'])]) / total * 100,
        'Not relevant & authoritative': len(group_df[(~group_df['is_relevant']) & (group_df['is_authoritative'])]) / total * 100,
        'Not relevant & not authoritative': len(group_df[(~group_df['is_relevant']) & (~group_df['is_authoritative'])]) / total * 100
    }

# Display as DataFrame
result_df = pd.DataFrame(result)
print(result_df)
```
